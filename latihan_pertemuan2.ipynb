```python
#latihan random number
import random
import math
size = 250
number = list()
ganjil = list()
power = list()
modus=list()
kamus=list()
tujuh_puluh = 0
min_number = 100
max_number = 0
total = 0

for i in range(size):
    number.append(random.randint(1,100))
for n in number:
    if n%2 != 0:
        ganjil.append(n)
for n in number:
    if n >70:
        tujuh_puluh=tujuh_puluh+1
for n in number:
    power.append(n**2)
for n in number:
    if n > max_number:
        max_number = n
for n in number:
    if n < min_number:
        min_number = n
        
sum_number= sum(number)
sum_power= sum(power)
power_sum = sum_number**2
variance = ((size*sum_power)-power_sum)/(size*size-1)
standard_deviation = math.sqrt(variance)

print (number)
print (ganjil)
print ("lebih dari 70:",tujuh_puluh)
print ("standar deviasi:",standard_deviation)
print ("min:",min_number)
print ("max:",max_number)
###Finding Mode
for i,n in enumerate(set(number)):
    value = n
    times = 0
    for num in number:
        if value == num:
            times = times + 1
    index= {"angka":value,"times":times}
    modus.append(index)
for dictionary in modus:
    if dictionary["times"]> total:
        total = dictionary["times"]
for dictionary in modus:
    if dictionary["times"] == total:
        print ("modus:",dictionary["angka"])
```
