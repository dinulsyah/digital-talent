# Analisa Data Realisasi Investasi Luar Negeri
# 1. Deskripsi Data
Dataset ini diambil langsung oleh BPS atau Badan Pusat Statistik dimana Dataset berisi mengenai jumlah dan realisasi investasi luar negeri per negara investor di Indonesia, BPS adalah Lembaga Pemerintah Non Departemen di Indonesia yang mempunyai fungsi pokok sebagai penyedia data statistik dasar, baik untuk pemerintah maupun untuk masyarakat umum, secara nasional maupun regional.

Data Terdiri dari Jumlahan data sebanyak 338 Data dan dari 5 kolom dimana sebagai berikut:<br>
A. nama_region: Nama Region <br>
B. nama_negara: Nama Negara <br>
C. tahun: Tahun <br>
D. jumlah_proyek: Jumlah Proyek <br>
E. nilai_investasi: Nilai Investasi (dalam juta USD)

# 2. Statistik Tentang Data Dan Visualisasi

```python
import pandas as pd
data_investasi = pd.read_csv("processed-data-realisasi-investasi-luar-negeri-per-negara-2000-2012.csv",delimiter = ',', encoding = "ISO-8859-1")
max_nilai = max(data_investasi['nilai_investasi'])
min_nilai = min(data_investasi['nilai_investasi'])
mean_nilai = data_investasi['nilai_investasi'].mean()
sum_nilai = data_investasi['nilai_investasi'].sum()
#Mencari Negara dengan Jumlah Nilai Investasi Tertinggi dan Terendah 
negara_tertinggi = data_investasi[data_investasi['nilai_investasi']==max_nilai]
negara_terendah = data_investasi[data_investasi['nilai_investasi']==min_nilai]
print("rata-rata nilai investasi keseluruhan:",mean_nilai,"\n")
print("jumlah nilai investasi keseluruhan:",sum_nilai,"\n")
print ("negara dengan nilai investasi tertinggi")
negara_tertinggi[['nama_negara','nilai_investasi']]
```

```python
print ("negara dengan nilai investasi terendah")
negara_terendah[['nama_negara','nilai_investasi']]
```

```python
max_proyek = data_investasi.jumlah_proyek.max()
min_proyek = data_investasi.jumlah_proyek.min()
mean_proyek = data_investasi.jumlah_proyek.mean()
sum_proyek = data_investasi.jumlah_proyek.sum()
proyek_tertinggi = data_investasi[data_investasi['jumlah_proyek']==max_proyek]
proyek_terendah = data_investasi[data_investasi['jumlah_proyek']==min_proyek]
print("rata-rata jumlah proyek keseluruhan tiap negara:",mean_proyek,"\n")
print("jumlah proyek keseluruhan:",sum_proyek,"\n")
print ("negara dengan proyek terbanyak")
proyek_tertinggi[['nama_negara','jumlah_proyek']]
```

```python
print ("negara dengan proyek terendah")
proyek_terendah[['nama_negara','jumlah_proyek']]
```

```python
data_investasi.groupby('tahun').mean()
```

```python
ax = data_investasi.groupby('tahun').mean()['jumlah_proyek']
ax = ax.plot(kind='bar', title ="Rata-Rata Proyek per Negara Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_investasi.groupby('tahun').mean()['nilai_investasi']
ax = ax.plot(kind='bar', title ="Rata-Rata nilai investasi per Negara Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah dalam (Million USD)", fontsize=12)
```

```python
ax = data_investasi.groupby('tahun').sum()['jumlah_proyek']
ax = ax.plot(kind='bar', title ="Jumlah Proyek Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_investasi.groupby('tahun').sum()['nilai_investasi']
ax = ax.plot(kind='bar', title ="Jumlah nilai investasi Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah dalam (Million USD)", fontsize=12)
```

```python
ax = data_investasi.groupby('nama_region').sum()['jumlah_proyek']
ax = ax.plot(kind='pie', title ="Jumlah Proyek Berdasarkan Region", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah Proyek", fontsize=12)
```

```python
ax = data_investasi.groupby('nama_region').sum()['nilai_investasi']
ax = ax.plot(kind='bar', title ="Jumlah nilai investasi Berdasarkan Region", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah dalam (Million USD)", fontsize=12)
```

```python
ax = data_investasi.groupby('nama_region').mean()['jumlah_proyek']
ax = ax.plot(kind='bar', title ="Jumlah Rata-Rata Proyek per Negara Berdasarkan Region", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah Proyek", fontsize=12)
```

```python
ax = data_investasi.groupby('nama_region').mean()['nilai_investasi']
ax = ax.plot(kind='bar', title ="Jumlah Rata-Rata Nilai investasi per Negara Berdasarkan Region", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah Proyek", fontsize=12)
```
