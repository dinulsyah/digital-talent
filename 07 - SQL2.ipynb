```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to many relationship

```python
run("SELECT customerNumber,CONCAT(firstName,' ',lastName) AS representative FROM customers c LEFT JOIN employees e ON e.employeeNumber=c.salesRepEmployeeNumber ORDER BY customerNumber")
```

```python
run("SELECT customerName,SUM(amount) AS total_payment FROM customers c INNER JOIN payments p ON c.customerNumber=p.customerNumber WHERE customerName='Atelier graphique'")
```

```python
run("SELECT orderDate,SUM(amount) FROM customers c INNER JOIN orders o ON c.customerNumber=o.customerNumber INNER JOIN payments p ON c.customerNumber=p.customerNumber GROUP BY orderDate ORDER BY orderDate")
```

```python
run("SELECT productName FROM orderdetails s INNER JOIN products p ON s.productCode=p.productCode INNER JOIN orders o ON s.orderNumber=o.orderNumber WHERE status NOT IN ('Shipped','Resolved') GROUP BY productName")
```

```python
run("SELECT customerNumber,sum(amount) FROM payments GROUP BY customerNumber")
```

```python
run("SELECT customerName,count(orderNumber) as total_order FROM customers c INNER JOIN orders o ON c.customerNumber=o.customerNumber WHERE customerName='Herkku Gifts'")
```

```python
run("SELECT CONCAT(firstName,' ',lastName) AS Employees_Boston FROM employees e INNER JOIN offices o ON e.officeCode=o.officeCode WHERE city='Boston'")
```

```python
run("SELECT customerNumber,SUM(amount) as total_payment FROM payments WHERE amount>100000 GROUP BY customerNumber ORDER BY amount DESC")
```

```python
run("SELECT COUNT(orderNumber) AS Value FROM orders WHERE status='On Hold'")
```

```python
run("SELECT customerName,COUNT(status) AS On_Hold FROM customers c INNER JOIN orders o ON c.customerNumber=o.customerNumber WHERE status='On Hold' GROUP BY customerName")
```

# Many to many relationship

```python
run("SELECT productName,orderDate FROM orderdetails s INNER JOIN products p ON s.productCode=p.productCode INNER JOIN orders o ON s.orderNumber=o.orderNumber WHERE status='Shipped'ORDER BY orderDate")
```

```python
run("SELECT productName,orderDate FROM orderdetails s INNER JOIN products p ON s.productCode=p.productCode INNER JOIN orders o ON s.orderNumber=o.orderNumber WHERE productName LIKE '%1940 Ford Pickup Truck%' ORDER BY orderDate DESC")
```

```python
run("SELECT DISTINCT(o.orderNumber),customerName FROM orders o INNER JOIN customers c ON o.customerNumber=c.customerNumber INNER JOIN orderdetails s ON o.orderNumber=s.orderNumber WHERE priceEach*quantityOrdered>25000")
```

```python
run("SELECT productCode FROM (SELECT o.productCode,COUNT(o.productCode) AS Appear FROM orderdetails o INNER JOIN products p ON o.productCode=p.productCode GROUP BY o.productCode) AS d WHERE appear >= (SELECT COUNT(DISTINCT orderNumber) AS orderNumber FROM orderdetails)") 
```

```python
run("SELECT DISTINCT productName AS product FROM products p INNER JOIN orderdetails o ON p.productCode=o.productCode WHERE priceEach<(MSRP*0.8)")
```

```python
run("SELECT DISTINCT productName FROM products p INNER JOIN orderdetails o ON p.productCode=o.productCode WHERE (priceEach-buyPrice)/buyPrice*100 > 100")
```

```python
run("SELECT orderNumber FROM orders WHERE orderDate IN (SELECT orderDate FROM orders WHERE DATENAME(dw,orderDate)='Monday')")
```

```python
run("SELECT productName FROM products WHERE productCode IN (SELECT productCode FROM orderdetails WHERE orderNumber IN (SELECT orderNumber FROM orders WHERE DATE_FORMAT(orderDate, '%a')='Mon'))")
```

```python
run("SELECT productName,quantityInStock FROM products WHERE productCode IN (SELECT productCode FROM orderdetails WHERE orderNumber IN (SELECT orderNumber FROM orders WHERE status='On Hold'))")
```
